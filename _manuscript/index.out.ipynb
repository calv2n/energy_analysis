## Dataset

PMJ Interconnection LLC (PMJ) is a regional transmission organization (RTO) operating an electric transmission system serving all or part of many states in the Eastern United States. It is considered the the largest independent power grid operator in North America ([PV Magazine USA 2025](#ref-pv2025pjm)).

The dataset we will use in our analysis contains the hourly and daily power consumption data from PJM. The dataset spans 2002 to 2018 and contains 145,336 hourly and 6,057 daily records. ([Kim 2025](#ref-wqg6-9917-25))
